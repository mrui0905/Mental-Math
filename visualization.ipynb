```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/large_multi.csv', header=None)
df
```

```python
coefficients = np.polyfit(df[0], df[1], 1)
m = coefficients[0]
b = coefficients[1]

plt.scatter(df[0], df[1])
plt.plot(df[0], m * df[0] + b, color='red')
plt.title('Score over time')
plt.xlabel('Attempt')
plt.ylabel('Number Correct')
plt.show()
```

```python
coefficients = np.polyfit(df[0], df[2], 1)
m = coefficients[0]
b = coefficients[1]



plt.scatter(df[0], df[2])
plt.plot(df[0], m * df[0] + b, color='red')
plt.title('Accuracy over time')
plt.xlabel('Attempt')
plt.ylabel('Percent Correct')
plt.show()
```
